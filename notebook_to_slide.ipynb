## Scatterplot(plt or sb) and talks about relationship

```python
df = pd.read_csv('fuel-econ.csv')
```

```python
df.info()
```

```python
df.fuelType.value_counts()
```

```python
df.corr().T
```

```python
plt.scatter(data = df, x='displ', y='cylinders');
```

```python
sb.regplot(data = df, x='displ', y='cylinders');
```

```python
plt.scatter(data = df, x='co2', y='comb');
```

```python
plt.scatter(data = df, x='displ', y='comb');
```

```python
sb.regplot(data = df, x='displ', y='comb');
```

```python
plt.figure(figsize=(12,6))
sb.heatmap(df.corr(), annot=True)
PLt.title('')
plt.xlabel()
plt.ylabel();
```

```python
color = sb.color_palette()[3]
plt.figure(figsize=(12,6))
sb.violinplot(data=df,x='VClass', y='co2', color=color);
```

```python
plt.figure(figsize=(12,6))
sb.boxplot(data=df,x='VClass', y='co2', color=color);
```

```python
sb.boxplot(data=df, y='co2');
```

```python
plt.figure(figsize=(12,6))
sb.boxplot(data=df,x='VClass', y='co2', hue='fuelType',color=color)
plt.legend(loc=(1,1));
```

```python
plt.figure(figsize=(12,6))
sb.violinplot(data=df,x='VClass', y='co2',hue='fuelType', color=color)
plt.legend(loc=(1,1));
```

```python
df['VClass'].value_counts().plot(kind='barh');
```

```python
plt.figure(figsize=(12,6))
sb.countplot(data=df, x='VClass', hue='fuelType',color=color)
```
